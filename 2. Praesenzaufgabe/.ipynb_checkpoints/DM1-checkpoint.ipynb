```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, make_scorer
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.feature_selection import f_classif
%matplotlib inline
```

```python
# check for data
if os.path.exists('image_parameters_smd_train.hdf5') is False:
    print('Download data from https://factdata.app.tu-dortmund.de/smd/small/image_parameters_smd_train.hdf5')
    
# read data and show data table
df_rf = pd.read_hdf('image_parameters_smd_train.hdf5')
df_rf
```

```python
# check for NaNs in any column
print(df_rf.columns.values[df_rf.isna().any()])
```

```python
# shuffle dataset
df_rf = df_rf.sample(frac=1).reset_index(drop=True)

# drop useless features and monte carlo truths
X = df_rf.drop(columns=['run_id', 'event_num',
                        'corsika_event_header_total_energy',
                        'corsika_run_header_particle_id'])

# create labels
y = (df_rf.corsika_run_header_particle_id == 1).values
```

```python
mu_s = np.mean(X[y])
mu_h = np.mean(X[~y])
var_s = np.std(X[y])**2
var_h = np.std(X[~y])**2

d = (mu_s - mu_h)**2 / (var_s + var_h)

d.sort_values(ascending=False)
```

```python
# feature selection to find 5 features with most information
F_val, p_val = f_classif(X, y)

feature_importances = pd.DataFrame(F_val,
                                   index = X.columns,
                                   columns=['importance']).sort_values('importance', ascending=False)
feature_importances
```

```python
# select the 5 best features
best_features = feature_importances['importance'][0:5].index.values
print(best_features)
X_5_features = df_rf[best_features]
```

```python
# create train and test sample
X_train, X_test, y_train, y_test = train_test_split(X_5_features, y,
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
# settings to test
settings_list = [
    {'n_estimators': 1, 'max_depth': None},
    {'n_estimators': 10, 'max_depth': None},
    {'n_estimators': 100, 'max_depth': None},
]
```

```python
# cross validation for the 3 settings
auc_list = []
for i, settings in enumerate(settings_list):

    # create classifier
    classifier = RandomForestClassifier(n_jobs=8, **settings)

    # compute auc in cross validation
    auc_values = cross_val_score(classifier, X=X_train, y=y_train, cv=5,
                                scoring=make_scorer(roc_auc_score))
    

    print('ROC AUC for settings {}: {:1.3f} +/- {:1.3f}'.format(
                    i, np.mean(auc_values), np.std(auc_values, ddof=1)))

    auc_list.append(auc_values)
    
    best_settings = settings_list[np.argmax(np.mean(auc_list, axis=1))]
print('best settings:', best_settings)
```

```python
classifier2 = DecisionTreeClassifier(max_depth=3)
auc_values2 = cross_val_score(classifier2, X=X_train, y=y_train, cv=5,
                                scoring=make_scorer(roc_auc_score))

print('ROC AUC for DecissionTree: {:1.3f} +/- {:1.3f}'.format(
                    np.mean(auc_values2), np.std(auc_values2, ddof=1)))
```

```python
classifier3 = MLPClassifier(alpha=10**(-5), max_iter=150)
auc_values3 = cross_val_score(classifier3, X=X_train, y=y_train, cv=5,
                                scoring=make_scorer(roc_auc_score))

print('ROC AUC for MLP: {:1.3f} +/- {:1.3f}'.format(
                    np.mean(auc_values3), np.std(auc_values3, ddof=1)))
```

```python
# create classifier
classifier = RandomForestClassifier(n_jobs=8, **best_settings)

# fit classifier
classifier3.fit(X_train, y_train)
```

```python
# get prediction
y_test_pred = classifier3.predict_proba(X_test)[..., 1]

# Get false positive and true positive rates
fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)

# calculate roc curve
roc_auc = auc(fpr, tpr)
print('ROC AUC:', roc_auc)

# plot roc curve
plt.rcParams.update({'font.size':18})
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(fpr, tpr, label='ROC (AUC = {:2.3f})'.format(roc_auc))
ax.plot([0, 1], [0, 1], linestyle='--', color='black',
        label='Random Guess')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('Receiver Operating Characteristic')
ax.legend(loc="lower right")
```

```python
# plot distributions
y_test_pred_proton = y_test_pred[y_test == 0]
y_test_pred_photon = y_test_pred[y_test == 1]

bins = np.linspace(0, 1, 50)
plt.rcParams.update({'font.size':18})
f, ax = plt.subplots(figsize=(8, 6))
ax.hist(y_test_pred_proton, bins=bins, label='Protons', histtype='step',
        fill=True, facecolor=[0.121, 0.466, 0.705, 0.15],
        edgecolor=u'#1f77b4')
ax.hist(y_test_pred_photon, bins=bins, label='Photons', histtype='step',
        fill=True, facecolor=[1., 0.498, 0.0549, 0.15],
        edgecolor=u'#ff7f0e')
ax.legend()
ax.set_xlabel('Classification Score')
ax.set_ylabel('Number of Events')
```
